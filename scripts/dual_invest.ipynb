```python
import datetime
import pandas as pd
"""
    ret[list] 的格式是  
    {
      id,  需要变成 index
      settleDate   1718092800000   => 需要改成 日期
      purchaseEndTime => 也需要改成日期
    }
""" 
df = pd.DataFrame(ret["list"])
df['settleDate'] = df['settleDate'].apply(lambda x: datetime.datetime.fromtimestamp(x / 1000).strftime('%Y-%m-%d'))
df['purchaseEndTime'] = df['purchaseEndTime'].apply(lambda x: datetime.datetime.fromtimestamp(x / 1000).strftime('%Y-%m-%d'))
df.set_index('id', inplace=True)

```

```python
# 比较测试


# 假设 df_a 和 df_b 是两个 DataFrame，包含 id 和 purchaseStatus 列
df_a = pd.DataFrame({'id': [1, 2, 3], 'purchaseStatus': ['pending', 'completed', 'completed']})
df_b = pd.DataFrame({'id': [1, 2, 3, 4], 'purchaseStatus': ['completed', 'pending', 'completed','new']})

# 使用 merge 函数比较两个 DataFrame
# df_merged = df_a.merge(df_b, on='id', suffixes=('_old', '_new'))
df_merged = df_a.merge(df_b, on='id', how="right", suffixes=('_old', '_new'),indicator=True)


# 筛选出 id 相同但是 purchaseStatus 不同的项目
df_result = df_merged[(df_merged['purchaseStatus_old'] != df_merged['purchaseStatus_new']) ]

print(df_merged)
# 打印结果
print("\n")
print(df_result)
```
